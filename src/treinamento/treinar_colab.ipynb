5.1 Criar Folds

````python
import json, os

# Descobre quais casos estão em imagesTr
imagestr_dir = f"{NNUNET_RAW}/Dataset001_BoneLesions/imagesTr"
cases = sorted([
    f.replace('_0000.nii.gz', '')
    for f in os.listdir(imagestr_dir)
    if f.endswith('_0000.nii.gz')
])
print(f'Casos encontrados ({len(cases)}): {cases}')

# Cria 5 folds ciclando pelos casos disponíveis
n = len(cases)
splits = []
for i in range(5):
    val_idx = i % n
    val   = [cases[val_idx]]
    train = [c for c in cases if c != cases[val_idx]]
    if not train:  # só 1 caso: usa ele mesmo para os dois
        train = val
    splits.append({'train': train, 'val': val})

splits_path = f"{NNUNET_PREPROCESSED}/Dataset001_BoneLesions/splits_final.json"
os.makedirs(os.path.dirname(splits_path), exist_ok=True)
with open(splits_path, 'w') as f:
    json.dump(splits, f, indent=2)

print('\n✅ splits_final.json criado:')
for i, s in enumerate(splits):
    print(f'  Fold {i}: treino={s["train"]}, val={s["val"]}')
````
